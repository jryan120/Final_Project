```python
def function (row):
    if row['P&I_percent'] > 7.2  and row['Date'] < 201829:
        return 'High Death Rate Early'
    elif row['P&I_percent'] > 7.2 and row['Date'] > 201829:
        return 'High Death Rate Late'
    
    
    
    
df_2['Death Rate']=df.apply(function, axis=1)

```
